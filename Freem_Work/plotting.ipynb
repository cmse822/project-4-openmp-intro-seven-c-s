```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Reading data
master_data = pd.read_csv("MMM_20x20.csv")
master_data = master_data.to_numpy()

# Removing first 20 entries of each timing to ignore first time load up costs
mean_data =  np.mean(master_data[21:-1, :], 0)
num_threads = np.array([1, 2, 4, 8, 12, 16, 20, 24])

plt.plot(num_threads, mean_data)
plt.xlabel("Thread Count")
plt.ylabel("Time to Compute (s)")
plt.title("Time to Compute MMM N=20")
plt.savefig("plot_N20.png")
```
